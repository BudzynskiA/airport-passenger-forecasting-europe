```python
import pandas as pd

# Wczytanie oczyszczonego pliku Excela
df = pd.read_excel("oczyszczone_dane_lotniska.xlsx")

# Wyświetlenie podstawowych informacji
print("➡️ Liczba wierszy i kolumn:", df.shape)

print("\n➡️ Pierwsze 5 wierszy danych:")
display(df.head())

# Sprawdzenie braków danych
print("\n➡️ Liczba braków danych w każdej kolumnie:")
print(df.isnull().sum())

# Typy danych
print("\n➡️ Typy danych:")
print(df.dtypes)

# Podstawowe statystyki opisowe (zachowane tylko do wyświetlenia)
print("\n➡️ Statystyki opisowe dla zmiennych liczbowych:")
display(df.select_dtypes(include='number').describe().T.round(2))

```
